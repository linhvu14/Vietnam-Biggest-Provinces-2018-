```python
# csv file: https://www.gso.gov.vn/default.aspx?tabid=714
# Vẽ biểu đồ 10 vùng có diện tích lớn nhất(2018)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('dientich2.csv',sep=';',header=1,index_col=0,skiprows=[3]).dropna()
df
```

```python
max_area = df['2018'].sort_values(ascending=True).tail(10)
max_area
```

```python
max_area.plot.bar()
plt.title('Đồ thị 10 tỉnh có diện tích lớn nhất Việt Nam năm 2018')
plt.legend(['Diện tích'])
```
